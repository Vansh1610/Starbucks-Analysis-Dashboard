```python
import pandas as pd

data=pd.read_csv('data.csv')

data=data[['County', 'ID']]
```

```python
data=data.groupby('County').count().reset_index()
data['number of starbucks']=data['ID']

```

```python
data['County']=data['County'].map(lambda x:x[:-6])
```

```python
data['County']=data['County'].map(lambda x:x[:-1])
```

```python
data.drop('ID',axis=1,inplace=True)
```

```python
data.to_csv('data.csv',index=False)
```

```python
import pandas as pd

data=pd.read_csv('income.csv')

data=data[['County', 'Median Household Income']]
```

```python
data['County']=data['County'].map(lambda x:x[:-6])
```

```python
data['Median Household Income']=data['Median Household Income'].map(lambda x:float(str(x).replace('$','').replace(',','')))
```

```python
type(data['Median Household Income'].iloc[0])
```

```python
data['Median Household Income'].fillna(round(data['Median Household Income'].mean()),inplace=True)
```

```python
data['County']=data['County'].map(lambda x:x[:-1])
```

```python
data.to_csv('income.csv',index=False)
```

```python
import pandas as pd

data=pd.read_csv('income_point.csv')

data=data[['id', 'coordinates']]
```

```python
data['coordinates']=data['coordinates'].map(lambda x:x.replace('POINT(','').replace(')','').split(' ')[1]+','+x.replace('POINT(','').replace(')','').split(' ')[0])
```

```python
data.to_csv('income_point.csv',index=False)
```
